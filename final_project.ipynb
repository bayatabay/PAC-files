```python
# תהליך ניתוח הנתונים:

# א. טעינה וסקירה ראשונית של הנתונים:

# טוענים את הנתונים לקובץ DataFrame
import pandas as pd

# קריאת קובץ CSV
df = pd.read_csv(r'C:\Users\user.LAPTOP-UJ6F34EO.000\food.csv')

# הצגת כמה שורות ראשונות כדי לראות את הנתונים
df.head()

# מבנה ה-DataFrame:
# מבנה ה-DataFrame כולל את כל הנתונים שהיו בקובץ ה-CSV. כל שורה מייצגת פריט מזון, וכוללת מספר עמודות המתארות את הערכים התזונתיים השונים.
# כל עמודה מייצגת משתנה אחר.

# הסבר על העמודות והנתונים:
# - Food_Name: שם המזון (למשל, "Apple", "Banana").
# - Calories: כמות הקלוריות למנה של המזון.
# - Protein: כמות החלבון במנת המזון.
# - Fat: כמות השומן במנת המזון.
# - Carbohydrates: כמות הפחמימות במנת המזון.
# - Fiber: כמות הסיבים במנת המזון.
# - Sugars: כמות הסוכר במנת המזון.
# - Vitamins: כמות הוויטמינים במנת המזון.
# - Minerals: כמות המינרלים במנת המזון.
```

```python
# טיפול בערכים חסרים:
# למילוי עם ממוצע:
df.loc[:, 'Data.Kilocalories'] = df['Data.Kilocalories'].fillna(df['Data.Kilocalories'].mean())

# או להסיר שורות עם ערכים חסרים:
df = df.dropna()

# סינון ערכים חריגים (Outliers):
Q1 = df['Data.Kilocalories'].quantile(0.25)
Q3 = df['Data.Kilocalories'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df = df[(df['Data.Kilocalories'] >= lower_bound) & (df['Data.Kilocalories'] <= upper_bound)]

# טרנספורמציות:
# המרת מספרים לקטגוריות (למשל חלבון ברמות שונות):
df['Protein_Category'] = pd.cut(df['Data.Protein'], bins=[0, 5, 10, 20, 50], labels=['Low', 'Moderate', 'High', 'Very High'])

# הצגת נתונים לאחר טיפול
df.head()
```

```python
# 1. סטטיסטיקות מסכמות
print("סטטיסטיקות מסכמות:")
print(df.describe())

# 2. סטטיסטיקות מסכמות עבור עמודות ספציפיות (למשל קלוריות, חלבון, סוכר)
print("\nסטטיסטיקות מסכמות עבור עמודת קלוריות:")
print(df['Data.Kilocalories'].describe())

print("\nסטטיסטיקות מסכמות עבור עמודת חלבון:")
print(df['Data.Protein'].describe())

print("\nסטטיסטיקות מסכמות עבור עמודת סוכר:")
print(df['Data.Sugar Total'].describe())

# 3. זיהוי ערכים קיצוניים - מינימום ומקסימום
print("\nערכים קיצוניים:")
min_value_kcal = df['Data.Kilocalories'].min()
max_value_kcal = df['Data.Kilocalories'].max()
print(f"Minimum value for Kilocalories: {min_value_kcal}")
print(f"Maximum value for Kilocalories: {max_value_kcal}")

min_value_protein = df['Data.Protein'].min()
max_value_protein = df['Data.Protein'].max()
print(f"Minimum value for Protein: {min_value_protein}")
print(f"Maximum value for Protein: {max_value_protein}")

min_value_sugar = df['Data.Sugar Total'].min()
max_value_sugar = df['Data.Sugar Total'].max()
print(f"Minimum value for Sugar: {min_value_sugar}")
print(f"Maximum value for Sugar: {max_value_sugar}")

# 4. ניתוח לפי קטגוריות (למשל, לפי קטגוריית מזון - 'Category')
print("\nסטטיסטיקות לפי קטגוריות (לפי 'Category'):")
category_stats = df.groupby('Category', observed=False)['Data.Kilocalories'].describe()
print(category_stats)

# 5. ניתוח לפי חלבון - יצירת קטגוריות
print("\nניתוח לפי חלבון - יצירת קטגוריות:")
df['Protein_Category'] = pd.cut(df['Data.Protein'], bins=[0, 5, 10, 20, 50], labels=['Low', 'Moderate', 'High', 'Very High'])
protein_category_analysis = df.groupby('Protein_Category', observed=False)['Data.Protein'].describe()
print(protein_category_analysis)

# 6. ניתוח לפי סוכר - יצירת קטגוריות
print("\nניתוח לפי סוכר - יצירת קטגוריות:")
df['Sugar_Category'] = pd.cut(df['Data.Sugar Total'], bins=[0, 5, 10, 20, 100], labels=['Low', 'Moderate', 'High', 'Very High'])
sugar_category_analysis = df.groupby('Sugar_Category', observed=False)['Data.Sugar Total'].describe()
print(sugar_category_analysis)

# 7. סטיית תקן עבור עמודות שונות
print("\nסטיית תקן עבור עמודות שונות:")
std_dev_kcal = df['Data.Kilocalories'].std()
print(f"Standard deviation for Kilocalories: {std_dev_kcal}")

std_dev_protein = df['Data.Protein'].std()
print(f"Standard deviation for Protein: {std_dev_protein}")

std_dev_sugar = df['Data.Sugar Total'].std()
print(f"Standard deviation for Sugar: {std_dev_sugar}")

# 8. ניתוח נוסף על המינונים של ויטמינים
print("\nניתוח סטטיסטי עבור ויטמינים (למשל, ויטמין A):")
df['Data.Vitamins.Vitamin A - RAE'] = pd.to_numeric(df['Data.Vitamins.Vitamin A - RAE'], errors='coerce')
vitamin_a_stats = df['Data.Vitamins.Vitamin A - RAE'].describe()
print(vitamin_a_stats)

```

```python
# 1. קיבוץ (Aggregation) נתונים לפי מימדים שונים

# קיבוץ לפי קטגוריית מזון ('Category') וחשיבת ממוצע של קלוריות, חלבון וסוכר
print("\nקיבוץ לפי קטגוריית מזון (Category) - ממוצע של קלוריות, חלבון וסוכר:")
category_aggregation = df.groupby('Category', observed=False)[['Data.Kilocalories', 'Data.Protein', 'Data.Sugar Total']].mean()
print(category_aggregation)

# קיבוץ לפי קטגוריית חלבון (Protein_Category) וחשיבת ממוצע של קלוריות
print("\nקיבוץ לפי קטגוריית חלבון (Protein_Category) - ממוצע של קלוריות:")
protein_category_aggregation = df.groupby('Protein_Category', observed=False)[['Data.Kilocalories']].mean()
print(protein_category_aggregation)

# קיבוץ לפי קטגוריית סוכר (Sugar_Category) וחשיבת ממוצע של חלבון
print("\nקיבוץ לפי קטגוריית סוכר (Sugar_Category) - ממוצע של חלבון:")
sugar_category_aggregation = df.groupby('Sugar_Category', observed=False)[['Data.Protein']].mean()
print(sugar_category_aggregation)

# 2. ניתוח מגמות על פני זמן (הנחה שאין עמודת תאריך)

# נניח שאין עמודת תאריך בקובץ, ולכן לא ניתן לבצע ניתוח מגמות על פני זמן. נמשיך לניתוחים לפי קטגוריות.

# 3. ניתוח מתקדם של סטטיסטיקות אחרות לפי קטגוריות

# ניתוח סטטיסטיקות (ממוצע, חציון, מינימום, מקסימום) לפי קטגוריית חלבון
print("\nסטטיסטיקות לפי קטגוריית חלבון (Protein_Category) - קלוריות:")
protein_category_stats = df.groupby('Protein_Category', observed=False)['Data.Kilocalories'].describe()
print(protein_category_stats)

# ניתוח סטטיסטיקות (ממוצע, חציון, מינימום, מקסימום) לפי קטגוריית סוכר
print("\nסטטיסטיקות לפי קטגוריית סוכר (Sugar_Category) - סוכר:")
sugar_category_stats = df.groupby('Sugar_Category', observed=False)['Data.Sugar Total'].describe()
print(sugar_category_stats)

# ניתוח לפי קטגוריית מזון (Category) - ממוצע של קלוריות, חלבון, וסוכר
print("\nסטטיסטיקות לפי קטגוריית מזון (Category) - קלוריות, חלבון וסוכר:")
category_stats = df.groupby('Category', observed=False)[['Data.Kilocalories', 'Data.Protein', 'Data.Sugar Total']].describe()
print(category_stats)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# סגנון עיצוב ברירת מחדל
plt.style.use('ggplot')  # סגנון שנמצא תמיד
sns.set(font_scale=1.1)

# בודקים אם יש ערכים חסרים
df = df.dropna(subset=['Data.Kilocalories', 'Data.Protein', 'Data.Sugar Total', 'Category'])

# תרשים 1: Pie chart - פילוח של 10 קטגוריות מזון עיקריות
top_categories = df['Category'].value_counts().nlargest(10)
plt.figure(figsize=(8, 8))
plt.pie(top_categories, labels=top_categories.index, autopct='%1.1f%%', startangle=140)
plt.title("Top 10 Food Categories")
plt.tight_layout()
plt.show()

# תרשים 2: Bar chart - ממוצע קלוריות לפי קטגוריה (10 מובילות)
avg_kcal = df.groupby('Category')['Data.Kilocalories'].mean().nlargest(10)
plt.figure(figsize=(10, 6))
avg_kcal.plot(kind='bar', color='salmon')
plt.title("Average Kilocalories by Category (Top 10)")
plt.ylabel("Kilocalories")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# תרשים 3: Barh chart - ממוצע חלבון לפי קטגוריה (10 מובילות)
avg_protein = df.groupby('Category')['Data.Protein'].mean().nlargest(10)
plt.figure(figsize=(10, 6))
avg_protein.plot(kind='barh', color='skyblue')
plt.title("Average Protein by Category (Top 10)")
plt.xlabel("Protein (g)")
plt.tight_layout()
plt.show()

# תרשים 4: Histogram - התפלגות רמות הסוכר
plt.figure(figsize=(8, 6))
plt.hist(df['Data.Sugar Total'], bins=30, color='orchid', edgecolor='black')
plt.title("Distribution of Total Sugar")
plt.xlabel("Sugar (g)")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()

# תרשים 5: Heatmap - מתאם בין ערכים תזונתיים נבחרים
nutrients = df[['Data.Protein', 'Data.Fat.Total Lipid', 'Data.Carbohydrate', 'Data.Sugar Total', 'Data.Kilocalories']]
plt.figure(figsize=(8, 6))
sns.heatmap(nutrients.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Between Nutritional Elements")
plt.tight_layout()
plt.show()
```
